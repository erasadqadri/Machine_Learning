```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from sklearn.datasets import load_boston
df = load_boston()
df
#data = input / independent feature
#target = output / dependent feature
```

```python
dataset = pd.DataFrame(df.data)
dataset.head()
```

```python
dataset.columns = df.feature_names
dataset.head()
```

```python
# independet and dependent features
X = dataset
y = df.target
```

```python
y
```

```python
# train test split
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.30, random_state = 42)
```

```python
X_train
```

```python
X_test
```

```python
y_train
```

```python
y_test
```

```python
#Standardizing the dataset
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler() #initializing SS
```

```python
X_train = scaler.fit_transform(X_train)
X_train
```

```python
X_test = scaler.transform(X_test)
X_test
```

```python
#linear regression
from sklearn.linear_model import LinearRegression
```

```python
#Cross Validation
from sklearn.model_selection import cross_val_score
```

```python
regression = LinearRegression()
regression.fit(X_train, y_train)
```

```python
#Mean Squared Error (difference between predicted and real value)
mse = cross_val_score(regression, X_train, y_train, scoring = 'neg_mean_squared_error', cv = 5)
mse
```

```python
np.mean(mse)
```

```python
#Prediction
reg_pred = regression.predict(X_test)
```

```python
reg_pred
```

```python
sns.distplot(reg_pred - y_test)
```

```python
from sklearn.metrics import r2_score
score = r2_score(reg_pred, y_test)
score
```
